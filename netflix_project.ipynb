## step 1 : import the libraries 

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load the data
df_netflix = pd.read_csv("netflix_titles.csv")
df_netflix
```

```python
# clean the data
df_netflix = df_netflix.dropna(subset = ['type','duration','release_year','rating','country','cast','title','director'])
df_netflix
```

```python
# reset the index 
df_netflix = df_netflix.reset_index(drop = True)
df_netflix 
```

```python
## plotting the type of graph 
type_counts = df_netflix['type'].value_counts()
print(type_counts)
print("Values : \n",type_counts.values)
print("Index : \n",type_counts.index)
```

```python
# plotting bar graph 
plt.figure(figsize = (8,6))
plt.bar(type_counts.index,type_counts.values,color = ["green",'yellow'],label = "Movies Vs Tv Show")
plt.xlabel("Type Of Shows On Netflix")
plt.ylabel("Count of Shows")
plt.title("Movies VS TV shows")
plt.legend()
# plt.yticks([1000,2000,3000,4000,5000])
plt.grid(True)
plt.tight_layout()
plt.savefig("Movies_vs_Tvshows.png")
plt.show()
```

```python
# plotting pie chart for rating 
rating_counts = df_netflix['rating'].value_counts()
print('Rating Counts Values: \n',rating_counts.values)
print('Index : ',rating_counts.index)

plt.figure(figsize = (10,6))
plt.pie(rating_counts.values,labels = rating_counts.index,autopct = "%1.2f%%",startangle = 90)
plt.title("Different shows Ratings")
plt.tight_layout()
plt.savefig('Ratings.png')
plt.show()
```

```python
movies_duration = df_netflix[df_netflix['type'] == 'Movie'].copy()
movies_duration['duration_int'] = movies_duration['duration'].str.replace(' min','').astype(float)
# movies_duration
```

```python
plt.figure(figsize = (8,8))
plt.hist(movies_duration['duration_int'],bins = 20,alpha = 0.9,hatch='o',color = 'gold',edgecolor = "black",label='Duration OF Movies')
plt.xlabel("Movies Duration in min")
plt.ylabel("Counts")
plt.legend()
# plt.ylim(0,1800)
plt.title("Count of Movies Duration")
plt.tight_layout()
plt.savefig('Movies_duration_counts.png')
plt.show()
```

```python
release_year = df_netflix['release_year'].value_counts().sort_index()
release_year

plt.figure(figsize = (6,7))
plt.scatter(release_year.index,release_year.values,color = 'purple',label = "Release Of Movies as per year")
plt.xlabel('Year ')
plt.ylabel('No. OF movies')
plt.title('Year VS NO. of Movies')
plt.legend()
plt.tight_layout()
plt.savefig('Year_vs_movies_release.png')
plt.show()
```

```python
# top 10 country graph

country_count = df_netflix['country'].value_counts().head(10)
plt.figure(figsize = (10,6))
plt.barh(country_count.index,country_count.values,color = 'gold',edgecolor = 'black',label = 'Top 10 Countries ')
plt.xlabel('No. Of Movies')
plt.ylabel('Country')
plt.title('Top 10 Countries in producing movies')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('Top_10_countries.png',bbox_inches = 'tight')
plt.show()
```

## movies vs tv shows per year graph by scatter

```python
content_per_year = df_netflix.groupby(['release_year','type']).size().unstack().fillna(0)

fig,ax = plt.subplots(1,2,figsize=(8,6))
ax[0].plot(content_per_year.index,content_per_year['Movie'],color = "gold")
ax[0].set_title('Movie Released per Year')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('No. of movies')

ax[1].plot(content_per_year.index,content_per_year['TV Show'],color = "gold")
ax[1].set_title('Tv show Released per Year')
ax[1].set_xlabel('Year')
ax[1].set_ylabel('No. of Tv show')

plt.suptitle('Movies and Tv shows released per year')
plt.savefig('content_per_year.png')
plt.tight_layout()
plt.show()
```
